```python
import pandas as pd
```

```python
# The attributes of the dataset are located in rows, which is not convenient, therefore data transpose was performed.
dataset = pd.read_csv('hazelnuts.txt', sep="\t").T
```

```python
dataset.head()
```

```python
# The index in the original dataset was also shuffled, therefore it was reset and put in order for further convenience. 
dataset.reset_index(drop=True, inplace=True)
```

```python
dataset.head()
```

```python
# The dataset did not have descriptive column names, therefore they were added manually based on specifications from the author.
dict = {0: 'length',
        1: 'width',
        2: 'thickness',
        3: 'surfacearea',
        4: 'mass',
        5: 'compactness',
        6: 'hardness',
        7: 'shelltopradius',
        8: 'watercontent',
        9: 'carbohydratecontent',
       10: 'variety'}

dataset.rename(columns=dict, inplace=True)
```

```python
dataset.head()
```

```python
# Saving preprocessed dataset.
dataset.to_csv('hazelnuts_preprocessed.csv', index=False)
```
